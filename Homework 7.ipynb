```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("personality20_labeled_oriented.csv")
df
```

```python
# Get the mean of each column
df_means = df.mean()
df_means
```

```python
# Demean each column
df = df.sub(df_means, axis = 'columns')
df
```

```python
# All columns now have a mean of approximately 0
df.mean()
```

```python
# Convert Dataframe to NumPy Array
A = np.array(df)
A
```

```python
U,S,V = np.linalg.svd(A)
```

```python
# Singular values of A
S
```

The first singular value of 226.87 is significantly larger than the others (the next largest singular value is 168.82). This means that this variable has more influence and accounts for more variance than the other variables. We do not need all of the questions to explain the variability in the data. 

There are a few other singular values that may help to account for more of the variability, but are not entirely necessary. We likely do not need all 5 traits of OCEAN to estimate the personality of someone because the large first singular value is able to explain the majority of the variability in the data.
