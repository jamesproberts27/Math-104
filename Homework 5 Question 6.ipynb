```python
# Part A

import numpy as np

years = np.array([1971,1972,1974,1978,1982,1985,1989,
                  1993,1997,1999,2000,2002,2003])
time = years - 1970

transistors = np.array([2250,2500,5000,29000,120000,275000,
                        1180000,3100000,7500000,24000000,42000000,
                        220000000,410000000])
logTransistors = np.log10(transistors)

import matplotlib.pyplot as plt

plt.scatter(time, logTransistors)
plt.xlabel("Time (Years since 1970)")
plt.ylabel("Log10(Number of Transistors)")
```

```python
# Part B
time = time.reshape(-1, 1)

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(time, logTransistors)
m = reg.coef_[0]
b = reg.intercept_

x = [0, 35]
y = [b, (b + 35 * m)]

plt.scatter(time, logTransistors)
plt.plot(x,y)
plt.xlabel("Time (Years since 1970)")
plt.ylabel("Log10(Number of Transistors)")
```

```python
# Part C
predict = b + (2015 - 1970) * m
predict = np.power(10,predict)
a = 4000000000

print("The prediction for the number of transistors in 2015 is", 
      np.round(predict / a, 3), "times as large." +
      "\nThis means Moore's slowed down a bit.")
```

```python
# Part D
moore = 2250 * np.power(2, (2015 - 1971) / 2)

plt.scatter(time, logTransistors)
plt.plot([0,45], [b, (b + 45 * m)], label = "Regression Line")
plt.plot([0,45], [np.log10(2250), np.log10(moore)], label = "Moore's law")
plt.xlabel("Time (Years since 1970)")
plt.ylabel("Log10(Number of Transistors)")
plt.legend()
```
