```python
import numpy as np
```

```python
def orthogonalProjection(A):
    U, S, V = np.linalg.svd(A, full_matrices = True)
    
    rank = 0
    for sv in S:
        if (sv > 0):
            rank += 1
    
    U1 = U[:,:rank]
    U1T = np.transpose(U1)
    
    P = np.matmul(U1,U1T)
    
    return P
```

```python
A = np.array([[1,2,3],[4,5,6],[1,0,8]])
print("A:\n", A)
ortho = orthogonalProjection(A)
print("Orthogonal Proj:\n", ortho)
print(np.matmul(A, ortho).round(10))
```

```python
A = np.array([[3,0,0],[0,2,0],[0,0,4], [1,1,1]])
print("A:\n", A)
ortho = orthogonalProjection(A)
print("Orthogonal Proj:\n", ortho)
print(np.matmul(ortho, A).round(10))
```
